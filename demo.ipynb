```python
from membership_function import MF
from inference_system import FuzzyWashSystem

import matplotlib.pyplot as plt
```

```python
for variable, mfs in MF.items():    
    fig, ax = plt.subplots()
    
    for item, mf in mfs.items():
        ax.plot(mf)
    
    ax.set_title(variable.name)
    ax.legend(mfs.keys(), loc='upper right')
    
    plt.show()
```

```python
# Create an instance of FuzzyWash
fuzzy_wash = FuzzyWashSystem()

# Define example input values
fabric_type_value = 7.5
dirt_level_value = 3.0
load_size_value = 8.0

wash_time_pred, water_temp_pred, agitation_intensity_pred = fuzzy_wash.predict(
    fabric_type_value, dirt_level_value, load_size_value,
)

# Print predicted values
print("Predicted Wash Time:", wash_time_pred)
print("Predicted Water Temperature:", water_temp_pred)
print("Predicted Agitation Intensity:", agitation_intensity_pred)

# Visualize the fuzzy sets
fuzzy_wash.visualize()
```
